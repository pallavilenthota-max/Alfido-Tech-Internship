```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load dataset
df = pd.read_csv('/content/website_wata.csv')
```

```python
#view dataset
df.head()
```

```python
#check datset information
df.info()
```

```python
#check missing values
df.isnull().sum()
```

```python
#summary statistics
df.describe()
```

```python
#craeting first chart
df.plot(x=df.columns[0], y=df.columns[1], kind='line')

plt.title('Website Visitors Trend')
plt.xlabel('Date')
plt.ylabel('Visitors')
plt.show()
```

```python
#creating second chart
traffic = df.groupby(df.columns[2])[df.columns[1]].sum()

traffic.plot(kind='pie', autopct='%1.1f%%')

plt.title('Traffic Sources')
plt.ylabel('')
plt.show()
```

```python
#creating third chart
device = df.groupby(df.columns[3])[df.columns[1]].sum()

device.plot(kind='bar')

plt.title('Device-wise Traffic')
plt.xlabel('Device')
plt.ylabel('Visitors')
plt.show()
```

## Insights

1. Website traffic changed over different time periods.

2. Some traffic sources generated higher visitor engagement.

3. Mobile/Desktop traffic comparison helps understand user behavior.

4. Website traffic analysis helps improve digital marketing strategies.
