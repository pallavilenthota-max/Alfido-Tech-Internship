```python
# load dataset
import pandas as pd

df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1')
```

```python
df.head()
```

```python
#check data information
df.info()
```

```python
#check missing values
df.isnull().sum()
```

```python
#summary statistics
df.describe()
```

```python
#create visualizations

#1.sales by region chart
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='bar')

plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
#2.profit by category chart
category_profit = df.groupby('Category')['Profit'].sum()

category_profit.plot(kind='pie', autopct='%1.1f%%')

plt.title('Profit by Category')
plt.ylabel('')
plt.show()
```

```python
#3.monthly sales trend chart
df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Month'] = df['Order Date'].dt.month

monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line', marker='o')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

## Insights

1. West region has the highest sales.

2. Technology category generated maximum profit.

3. Sales increased during the end of the year.
